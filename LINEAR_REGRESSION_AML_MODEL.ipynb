```python
# IMPORT LIBRARIES

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#LOAD DATASET

data = pd.read_excel('country.xlsx')
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
#PRINT FIRST 5 ROW

data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
import pickle
from sklearn.preprocessing import LabelEncoder

# Assuming 'payment' is a categorical column in your DataFrame
columns = ['Code', 'Name', 'Description', 'isocode3digit', 'Riskscore', 'Optional_document', 'rating_source', 'risk_assessment']

# Initialize LabelEncoders
label_encoders = {}

# Fit and transform the data
for col in columns:
    label_encoder = LabelEncoder()
    data[col] = data[col].astype(str)  # Convert all values to strings
    type_encoded = label_encoder.fit_transform(data[col])
    data[col+'_encoded'] = type_encoded
    label_encoders[col] = label_encoder

# Save the encoders
for col, encoder in label_encoders.items():
    with open(col+'_encoder.pkl', 'wb') as f:
        pickle.dump(encoder, f)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```
